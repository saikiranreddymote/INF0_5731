Question 2 (20 points): Write python code to extract these features you discussed above. You can collect a few sample text data for the feature extraction. 

```python
# You code here (Please add comments in the code):

import nltk

from bs4 import BeautifulSoup as bs
import pandas as pd
import requests as rq
import matplotlib.pyplot as plt
import nltk
from wordcloud import WordCloud
from nltk.corpus import stopwords

#collecting sample text data from IMDb "KGF2 movie reviews"

url = 'https://m.imdb.com/title/tt7286456/reviews?ref_=tt_ov_rt'

#Storing the reviews into a dataframe
df = pd.DataFrame(columns=['Review'])
req = rq.get(url).text
soup = bs(req,'html.parser')

#collecting reviews as text
reviews_text = soup.find_all('div', attrs={'class' : 'text show-more__control'})

#taking a empty list initially
review_list = []
for i in range(len(reviews_text)):
    review_list.append(reviews_text[i].get_text())

#assigining list of reviews to the data frame    
df['Review'] = review_list

df['word_count'] = df['Review'].apply(lambda x: len(str(x).split(" ")))
df[['Review','word_count']].head()

```

```python
#character count

df['char_count'] = df['Review'].str.len() ## including spaces
df[['Review','char_count']].head()
```

```python
# Getting Stopwords

from nltk.corpus import stopwords

import nltk
nltk.download('stopwords')
stop = stopwords.words('english')

df['stopwords'] = df['Review'].apply(lambda x: len([x for x in x.split() if x in stop]))
df[['Review','stopwords']].head()
```

```python
#hashtags in the reviews
df['hastags'] = df['Review'].apply(lambda x: len([x for x in x.split() if x.startswith('#')]))
df[['Review','hastags']].head()
```

```python
#numerics in the reviews
df['numerics'] = df['Review'].apply(lambda x: len([x for x in x.split() if x.isdigit()]))
df[['Review','numerics']].head()
```

```python
#Upper characters in the reviews
df['upper'] = df['Review'].apply(lambda x: len([x for x in x.split() if x.isupper()]))
df[['Review','upper']].head()
```
